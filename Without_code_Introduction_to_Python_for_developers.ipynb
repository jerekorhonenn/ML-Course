### Python Data Types

Python has a number of built-in data types that you can use to store different kinds of data. Here are some of the most commonly used data types in Python:

- int: integer numbers (e.g. 42)
- float: floating-point numbers (e.g. 3.14)
- bool: boolean values (True or False)
- str: string values (e.g. "Hello, World!")
- list: list of values (e.g. [1, 2, 3])
- tuple: tuple of values (e.g. (1, 2, 3))
- dict: dictionary of key-value pairs (e.g. {'name': 'Alice', 'age': 42})

Python does not have a strict typing system like many other programming languages, variables don't have to be explicitly declared with a specific type, also this allows more flexibility and less verbosity in the code.

#### Some cases when to use casting

Casting is often used in situations where the data type of a variable needs to be changed to match the expected data type of a function or operator. Here are a few examples of when casting might be used:

**When performing mathematical operations**: Some operations in Python, such as division, return a float by default. However, if you want to store the result as an integer, you will need to cast it using the `int()` function.

**When working with user input**: In many cases, user input is read as a string, even if it represents a number. To perform arithmetic operations or comparisons on this input, you will need to cast it to the appropriate data type.

## Control Flow

Python, like many programming languages, has several control flow constructs that allow you to control the flow of execution in your code.

`if statements`: Used to conditionally execute code. The general syntax is as follows:

```
if condition:
    # code to execute if condition is True
else:
    # code to execute if condition is False

```

`for loops`: Used to iterate over a sequence (such as a list or a string) and execute a block of code for each item in the sequence. 

```
for item in sequence:
    # code to execute for each item in the sequence
```

`while loops`: Used to repeatedly execute a block of code as long as a certain condition is True.


```
while condition:
    # code to execute while condition is True
```

`break and continue statements`: Used to control the flow of a loop. `break` is used to exit a loop early, while `continue` is used to skip the current iteration of a loop and continue with the next iteration.

## Functions

In Python, a function is a block of code that performs a specific task and may return a result. Functions help break down a program into smaller and modular chunks of code, making it easier to understand, test and debug.

To define a function in Python, you use the def keyword, followed by the name of the function, a set of parentheses, and a colon. The code block inside the function is indented, as with other control flow constructs.

Functions can also return a value, which can be stored in a variable or used as part of an expression.

### Default values for arguments

You can provide default values for function arguments, which means that if the argument is not passed when the function is called, the default value will be used instead.

### args, kwargs

In Python, function arguments can be passed in the order they are defined (positional arguments), by explicitly mentioning the name of the argument while calling the function(keyword arguments) or by using a combination of both.

### Classes

A class is a blueprint for creating objects (i.e. instances of a class), which can have properties and methods. A class is defined using the `class` keyword, followed by the name of the class and a colon. The code block inside the class is indented, as with other control flow constructs.

```
class MyClass:
    """This is a simple class"""
    pass

```

To create an object of a class, you use the class name followed by parentheses:

`obj = MyClass()`

Inside the class, you can define class variables, which are variables that are shared by all objects of the class, and instance variables, which are specific to each object. The `self` parameter is used to refer to the current instance of the class.

Classes can have methods, which are functions that are defined inside a class and are used to perform actions on the class's objects.

Inheritance is another important feature in OOP, where a class can inherit properties and methods from a base class. This allows for the creation of new classes that are built on existing ones.

## Libraries for Data Science/Machine Learning

`NumPy`: A library for working with large arrays and matrices of numerical data, and performing mathematical operations on them. It provides functionality for working with arrays and matrices, as well as a large library of mathematical functions.

````python
import numpy as np

# Creating an array
a = np.array([1, 2, 3])

# Perform mathematical operations on arrays
b = np.array([4, 5, 6])
print(a + b) # Output: [5, 7, 9]
print(a * b) # Output: [4, 10, 18]

# Perform mathematical operations on array elements
print(np.sin(a)) # Output: [0.84147098, 0.90929743, 0.14112001]
````
